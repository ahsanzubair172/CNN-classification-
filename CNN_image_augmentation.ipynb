```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array

# Load original image
original_img = load_img('/content/cat.4198.jpg', target_size=(150, 150))
img_array = img_to_array(original_img)
img_array = img_array.reshape((1,) + img_array.shape)

# Show 3 variations for a given transformation
def show_transform(title, datagen):
    gen = datagen.flow(img_array, batch_size=1)

    plt.figure(figsize=(10, 4))
    plt.suptitle(title, fontsize=16)

    # Original image
    plt.subplot(1, 4, 1)
    plt.imshow(original_img)
    plt.title('Original')
    plt.axis('off')

    # Generate 3 transformed versions
    for i in range(3):
        transformed_img = next(gen)[0]

        # Normalize if needed
        if transformed_img.max() > 1.0:
            transformed_img = transformed_img / 255.0

        plt.subplot(1, 4, i+2)
        plt.imshow(transformed_img)
        plt.title(f'Augmented {i+1}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# 1. Rescale
datagen_rescale = ImageDataGenerator(rescale=1./255)
show_transform("Rescale (Normalization)", datagen_rescale)

# 2. Shear
datagen_shear = ImageDataGenerator(shear_range=0.8)
show_transform("Shear Transformation", datagen_shear)

# 3. Zoom
datagen_zoom = ImageDataGenerator(zoom_range=0.7)
show_transform("Zoom Transformation", datagen_zoom)

# 4. Horizontal Flip
datagen_hflip = ImageDataGenerator(horizontal_flip=True)
show_transform("Horizontal Flip", datagen_hflip)

# 5. Vertical Flip
datagen_vflip = ImageDataGenerator(vertical_flip=True)
show_transform("Vertical Flip", datagen_vflip)
```

```python

import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array

# Load original image
original_img = load_img('/content/cat.4198.jpg', target_size=(150, 150))
img_array = img_to_array(original_img)
img_array = img_array.reshape((1,) + img_array.shape)

# Show 3 variations for a given transformation
def show_transform(title, datagen):
    gen = datagen.flow(img_array, batch_size=1)

    plt.figure(figsize=(10, 4))
    plt.suptitle(title, fontsize=16)

    # Original image
    plt.subplot(1, 4, 1)
    plt.imshow(original_img)
    plt.title('Original')
    plt.axis('off')

    # Generate 3 transformed versions
    for i in range(3):
        transformed_img = next(gen)[0]

        # Normalize if needed
        if transformed_img.max() > 1.0:
            transformed_img = transformed_img / 255.0

        plt.subplot(1, 4, i+2)
        plt.imshow(transformed_img)
        plt.title(f'Augmented {i+1}')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Show manual vertical flip
def show_vertical_flip_manual(img_tensor):
    img = img_tensor[0]  # remove batch dim
    flipped_img = np.flipud(img)

    plt.figure(figsize=(6, 3))
    plt.suptitle("Vertical Flip (Manual)", fontsize=14)

    plt.subplot(1, 2, 1)
    plt.imshow(img.astype('uint8'))
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(flipped_img.astype('uint8'))
    plt.title("Flipped")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# 1. Rescale
datagen_rescale = ImageDataGenerator(rescale=1./255)
show_transform("Rescale (Normalization)", datagen_rescale)

# 2. Shear
datagen_shear = ImageDataGenerator(shear_range=0.8)
show_transform("Shear Transformation", datagen_shear)

# 3. Zoom
datagen_zoom = ImageDataGenerator(zoom_range=0.7)
show_transform("Zoom Transformation", datagen_zoom)

# 4. Horizontal Flip
datagen_hflip = ImageDataGenerator(horizontal_flip=True)
show_transform("Horizontal Flip", datagen_hflip)

# 5. Vertical Flip (Manual – Always works)
show_vertical_flip_manual(img_array)
```
